### Task 1 

`Simple Monte Carlo` is an algorithm that relies on repeated random sampling to obtain numerical results. it is mainly used to estimate the probability or expected value when a solution is difficult or imposible to derive

The method works by **simulating** a random process, a large number of times (**$n$**). By aggregating the results of  the independant simulations, The Law of Large numbers ensures that the empirical mean of the samples converge to the true theoretical mean (expected value) of the system  $n \to \infty$.

The Following **assumptions** must be made so that `Simple Monte Carlo` can provide valid estimates

- The Samples generated must be independent of each other meaning that the outcome of one simulation does not affect the next and are drawn from the same probability distribution.

- it is also assumed that the random variable has a finite variance meaning not $\infty$. This assumption enables the calculation of confidence intervals and error estimates

#### Mean Estimation

The Theoretical mean of a random variable is estimated by calculating the sample mean of the results obtained from the $n$ simulations.

The Formula for estimating the mean is as follows; if $X_1,X_2,...,X_n$ are the results of $n$ independent simulations than the estimated mean ($\hat{\mu}_n$) is given by:

$$\hat{\mu}_n = \frac{1}{n} \sum_{i=1}^{n} X_i$$

where $\hat{\mu}_n$ is the estimated mean, $n$ is the number of simulations, $X_i$ is the result of the $i$-th simulation.


#### Error Analysis

- An estimator is considered **unbiased** if its expected value is equal to the true parameter it is estimating. In `Simple Monte Carlo`, the sample mean is an unbiased estimator of the true population mean. This means that is a simulation of an entire set is run $n$ simulations infinite times, than the average of our estimates would be equal to the true answer

- The variance of the monte carlo estimator measures how much the estimate fluctuates around the true mean. it is derived from the variance of a single simulation divided by the sample size. **For Example** to reduce the varience of the estimate by a factor of `10` , the number of simulations $n$ need to increase by a factor of `100`

- For `Monte Carlo`, the RMSE is equal to the standard error of the mean. it can be estimated using the sample standard deviation ($s$)

$$RMSE = \sqrt{Var(\hat{\mu}_n)} = \frac{\sigma}{\sqrt{n}} \approx \frac{s}{\sqrt{n}}$$

where $s$ is the sample standard deviation of the results, $n$ is the number of simulations
